```python
import pandas as pd
dataset = pd.read_csv('a1_RestaurantReviews_HistoricDump.tsv', delimiter = '\t', quoting = 3)
```

```python
dataset.shape
```

```python
dataset.head()
```

```python
from keras.preprocessing.text import Tokenizer
tokenizer = Tokenizer()
```

```python
docs=dataset["Review"]
```

```python
tokenizer.fit_on_texts(docs)
```

```python
tokenizer.word_index
```

```python
len(tokenizer.word_index)
```

```python
sequences = tokenizer.texts_to_sequences(docs)
sequences
```

```python
docs
```

```python
import numpy as np
import statistics
statistics.median((docs.apply(len)).values)
```

```python
from keras.utils import pad_sequences
sequences = pad_sequences(sequences,padding='post',maxlen=51)
sequences
```

```python
from keras import Sequential
from keras.layers import Dense,SimpleRNN,Embedding,Flatten
```

```python
model = Sequential()
model.add(Embedding(1932,2,input_length=51))
model.add(SimpleRNN(32,return_sequences=False))
model.add(Dense(1, activation='sigmoid'))
```

```python
X=sequences
```

```python
Y=dataset['Liked']
```

```python
X
```

```python
Y=Y.to_numpy()
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['acc'])
model.fit(X,Y,epochs=5)
```
